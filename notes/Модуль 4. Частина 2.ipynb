````python
import random
import pathlib
import colorama
import time
from tqdm import tqdm

jokes = ["Why don't scientists trust atoms? Because they make up everything!",
    "Parallel lines have so much in common... it's a shame they'll never meet.",
    "I'm reading a book about anti-gravity. It's impossible to put down!",
    "Did you hear about the mathematician who's afraid of negative numbers? He'll stop at nothing to avoid them.",
    "Why did the scarecrow win an award? Because he was outstanding in his field!", ]

random_value = '()+, -0123456789;=@ABCDEF HIJKL       MNOPQRSTUVWXYZ` abcdefghijklmnopqrstuvwxyz'


def get_random_joke(directory):
    current_dir = pathlib.Path(f'{directory}')
    try:
        with open(current_dir / "jokes .txt", "r", encoding="utf-8") as file:
            jokes = file.readlines()
            return colorama.Fore.WHITE + random.choice(jokes).strip()
    except FileNotFoundError:
        print(colorama.Fore.RED + "Unexpected error!!!")
        # return hard_reset_imitation()
        # return cli_progress_test('Analysing error', 100)
        cli_progress_test('Analysing error', 100)
        return need_to_connect_ai()


def training_ai_imitation():
    for _ in tqdm(range(100)):
        time.sleep(0.1)


def cli_progress_test(text_in_message, end_val, bar_length=20):
    for i in range(0, end_val):
        time.sleep(0.01)
        percent = float(i) / end_val
        hashes = '#' * int(round(percent * bar_length))
        spaces = ' ' * (bar_length - len(hashes))
        print("\r{0}: [{1}] {2}%".format(text_in_message, hashes + spaces, int(round(percent * 100))))


def need_to_connect_ai():
    print(colorama.Fore.YELLOW + """Creating AI to solve issue with jokes\n\
AI trying to understand what cause Exception\n\
Not enough CPU!!! Install additional CPU""")
    cli_progress_test(colorama.Fore.GREEN + 'Install CPU', 100)
    print(colorama.Fore.YELLOW + """CPU installed recheck code\n\
AI couldn't detect where issue occur\n\
Updating AI to generate jokes\n\
Start training on world wide jokes via net""")
    training_ai_imitation()
    print(f"Training finished\nIteration 1: {''.join(random.choices(random_value, k=100))}\n\
Iteration 2: {' '.join(random.choices(' '.join(jokes).split(' '), k=16))}")
    final_joke = random.choice(jokes)
    print(f"Iteration 3: {final_joke}")
    return colorama.Fore.WHITE + final_joke


````
